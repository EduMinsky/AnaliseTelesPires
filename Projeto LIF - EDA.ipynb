```python
#Criando coluna com os Estados de cada Pedido
df_estados = df[df.cod_imovel!=' ']
df_estados['UF'] = df[df.cod_imovel!=' '].cod_imovel.str[:2]
df_estados

df_sem = df[df.cod_imovel==' ']
df_sem['UF']= 'Indisp.'
df_new = pd.concat([df_estados,df_sem])
df_new
```

```python
#Plotando essa informação
sns.set_palette("PuBuGn_d")
sns.set(rc={"figure.figsize":(10, 10)})
sns.barplot(data=df_new.UF.value_counts().to_frame().reset_index(), x="index", y="UF")
```

```python
#Passando as colunas de data para datetime
df_new['dat_valida_Date'] = pd.to_datetime(df_new['dat_valida'])
df_new['dat_vali00_Date']= pd.to_datetime(df_new['dat_vali00'])
df_new['dat_valida_Date_YEAR'] = df_new['dat_valida_Date'].dt.year
df_new['dat_vali00_Date_YEAR']= df_new['dat_vali00_Date'].dt.year
```

```python
#Calculando o tempo de duração do Pedido
df_tempo=[]
for i in range(len(df_new)):
    df_tempo.append(df_new['dat_vali00_Date'].iloc[i] - df_new['dat_valida_Date'].iloc[i])
df_new['Tempo_Autorizacao'] = df_tempo
df_new['Tempo_Autorizacao'] = df_new.Tempo_Autorizacao/ np.timedelta64(1, 'Y')
```

```python
df_uas = df_new.query('nom_ativid=="Uso Alternativo do Solo"')

sns.set_palette("PuBuGn_d")
sns.set(rc={"figure.figsize":(10, 10)})
sns.barplot(data=df_uas.UF.value_counts().to_frame().reset_index(), x="index", y="UF")
df_uas.UF.value_counts()
```

```python
df_uas['st_area_sh'] = df_uas['st_area_sh'].str.replace(',', '.').astype(float)
df_uas.asd = df_uas.asd.str.replace(',', '.').astype(float)
```
